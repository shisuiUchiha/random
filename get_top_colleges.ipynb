```python
# Query google and find the top colleges list
from googlesearch import search
import pandas as pd
import requests
from bs4 import BeautifulSoup
import re


```

```python
class GoogleQuery:
    def __init__(self, query):
        self.query = query
    
    def fetch_urls(self):
        print("Fetching Results:")
        df = {}
        for idx, result in enumerate(search(self.query, num_results=10), start=1):
            df[idx] = result
        return pd.DataFrame(df.items(), columns=['rank', 'result'])
    
```

```python
class bsparse:
    def __init__(self, url):
        self.url = url
    
    def parse_urls(self):
        headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.1; WOW64; rv:50.0) Gecko/20100101 Firefox/50.0'}
        response = requests.get(self.url, headers = headers, timeout=5)
        soup = BeautifulSoup(response.content, "html.parser")
        heading = soup.find_all('h3', )
        college_list = []
        for head in heading:
            try:
                val = head.string
                college_list.append(head.string)
            except:
                val = None
                college_list.append("error")            
        return college_list
```

```python
cities = ["Hyderabad","Bangalore","Kolkata","Delhi","Mumbai","Chennai","Kerala","Karnataka","Telengana","Andhra Pradesh","Shillong","Guwahati","Assam","Meghalaya","Tripura","Sikkim"]
clg_list = []
for city in cities:
    a = GoogleQuery(f"Top Commerce Colleges in {city}")
    df = a.fetch_urls()
    df = df[df['result'].str.contains('iide')]
    clg_list.append(df)
final_df = pd.concat(clg_list)
final_df = final_df.drop_duplicates(subset = ["result"])
final_df
```

```python
final_df
```

```python
final_clg = {}
for link in final_df["result"].to_list():
    b = bsparse(link)
    l = b.parse_urls()
    l = [i for i in l if i]
    pattern = r'[0-9]*[.]'
    l = [i for i in l if re.match(pattern,i[0:3])]
    final_clg[link.split("-")[-1]] = l
final_colleges = pd.DataFrame(dict([ (k,pd.Series(v)) for k,v in final_clg.items() ]))
```

```python
def get_email_contacts(url):
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.1; WOW64; rv:50.0) Gecko/20100101 Firefox/50.0'}
    response = requests.get(url, headers = headers, timeout=5)
    soup = BeautifulSoup(response.content, "html.parser")
    mailtos = soup.select('a[href]')
    all_emails = []
    for i in mailtos:
        if "mailto" in i["href"]:
            all_emails.append((i["href"].replace("mailto:","")))
    return all_emails
```

```python
contact_df = pd.DataFrame(columns = ["city","college","url","emails"])
i=0
import numpy as np
for city in final_colleges.columns:
    clgs = final_colleges[city]
    clg_list = clgs.dropna().to_list()
    clg_list =[re.sub(r'[0-9.]+', '', s) for s in clg_list]
    for clg in clg_list:
        a= GoogleQuery(f"{clg} placement contact")
        import time
        time.sleep(15)
        df = a.fetch_urls()
        try:
            url = df["result"].iloc[0]
            emails = get_email_contacts(url)
        except:
            url = None
            emails = None
        contact_df.loc[i] = [city,clg,url,emails]
        i=i+1
print(contact_df)
        
    


        

        
```
